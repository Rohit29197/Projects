```python
from os import listdir
x=[]   
y=[]
for category in listdir(r"C:\Users\Rohit/20_newsgroups"): 
    for document in listdir(r"C:/Users/Rohit/20_newsgroups/"+category): 
        with open(r"C:/Users/Rohit/20_newsgroups/"+category+'/'+document, "r") as f:
            x.append(f.read())
            y.append(category)
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
stop= stopwords.words('english')
punctuations=list(string.punctuation)
stop=stop+punctuations
lemmatizer= WordNetLemmatizer()
```

```python
def data_cleaning(data):
   clean_data=[]
   for i in range (len(x)):
      words=word_tokenize(x[i].lower())
      #remove_stopwords=[w for w in words if not w in stop]
      clean=[]
      for w in words:
         if w.lower() not in stop:
            pos=pos_tag([w])
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            clean.append(clean_words.lower())
      clean_data.append(clean)
   return clean_data
```

```python
clean_data = data_cleaning(x)
documents=[" ".join(document)for document in clean_data]
documents[10]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
x_train,x_test,y_train,y_test = train_test_split(documents,y)
count_vector = CountVectorizer (max_features=5000)
x_train_features = count_vector.fit_transform(x_train)
x_test_features = count_vector.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
alg = MultinomialNB()
alg.fit(x_train_features,y_train)
y_predict=alg.predict(x_test_features)
print(classification_report(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```
